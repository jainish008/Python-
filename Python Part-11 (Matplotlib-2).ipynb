```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = [1, 2, 3, 4, 5, 6, 7]

sleeping = [7, 7, 7, 7, 7, 9, 9]
eating = [2, 2, 2, 2, 2, 3, 3]
college_time = [8, 9, 8, 9, 8, 2, 2]
playing = [2, 2, 2, 2, 2, 4, 4]
self_learning = [1, 1, 1, 1, 1, 4, 4]

#For stackplot and colors in it
plt.stackplot(days, sleeping,eating,college_time,playing,self_learning, colors=['m','b','r','k','c'])

plt.xlabel('Days in Week')
plt.ylabel('Hours of daily activity')
plt.title('My Daily Life during College')
plt.show()
```
