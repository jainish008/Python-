```python
import pandas as pd
import numpy as np
```

Create a Data 
5 friends names and birthyear

```python
#create a variable name  friends
friends = ["JAINISH","Nikita","Yash","Kapil","Sam"]
birthyear = [1999, 1999, 2000, 1994, 1998]
```

To merge these two lists together we will use the zip function.

```python
FriendSet = list(zip(friends,birthyear))
FriendSet
```

```python
#df will be a DataFrame object. You can think of this object holding the contents of the BabyDataSet in a 
#format similar to a sql table or an excel spreadsheet. Lets take a look below at the contents inside df.
df = pd.DataFrame(data = FriendSet, columns=["FriendName","BirthYear"])
df
```

```python
df.to_csv('Friend Birth Year.csv',index=False,header=False)
```

```python
#sorting of values
nba = pd.read_csv("nba.csv")
```

```python
nba
```

```python
nba.head(10)
```

```python
#sort the nba by name and by default it will be ascending order
nba.sort_values("Name")
```

```python
#sorting by name, but in descending order 
nba.sort_values("Name", ascending=False)
```

```python
#sorting values by salary in ascending order
nba.sort_values("Salary", ascending=True)
```

```python
#sorting the data by columns 
nba.sort_values(["Team","Name"], ascending=True)
#in this first it sort by team in ascending order and then it will sort
#by name in ascending order
```

```python
#now I want three columns team, name, salary
#Team should be in descening order
#Name and Salary should be in ascending order
nba.sort_values(["Team","Name","Salary"], ascending=[False,True,True])
```

```python
#sorting of index
nba.sort_index()
```

```python
#reverse the index
nba.sort_index(ascending=False)
```
