```python
#Importing lib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Plot a basic graph in matplotlib
#in which [1,2,3] are x axis 
#in second [9,4,5] are y axis 
#as you can see in graph 9 from which line is going downward
#as in x axis there 2 which is in focus 
plt.plot([1,2,3],[9,4,5])
plt.show()
```

```python
#Another example changing only 2 in axis and 9 in y axis to know the difference 
#and understand the positions in graph
plt.plot([1,3,2.5],[7,4,10])
plt.show()
#As we can see difference where 10 is making the y axis big.
```

Now all the above graphs where displaying and not telling what is the data
So, for the better understanding of the data in the graph we need to use 
titles labels legend which are part of matplotlib so that client or any other
person read or understand data in the graph that has been presented by you

```python
#Below we are listing three movies of nolan on x axis 
#Box-Office earning of that movie on y axis 
x = ["Inception","Interstellar","Dunkirk"]
y = [82.2, 67.75, 56.69]
```

```python
plt.plot(x,y)
plt.xlabel("Movie List")
plt.ylabel("Box-Office Collection in USD")
plt.title("Christopher Nolan Movie Earning")
plt.show()
```

```python
#Now same graph in bar-chart format and not line
plt.bar(x,y)
plt.xlabel("Movie List")
plt.ylabel("Box-Office Collection in USD")
plt.title("Christopher Nolan Movie Earning")
plt.show()
```

```python
#Now same graph in scatter-plot format
plt.scatter(x,y, color="K")
plt.xlabel("Movie List")
plt.ylabel("Box-Office Collection in USD")
plt.title("Christopher Nolan Movie Earning")
plt.show()
```

```python
#In scatter option we have different option of marker(which is circle in above graph)
#Below is scatter graph with different marker option
plt.scatter(x,y, color="K", marker="v")
plt.xlabel("Movie List")
plt.ylabel("Box-Office Collection in USD")
plt.title("Christopher Nolan Movie Earning")
plt.show()
```

```python
plt.scatter(x,y, color="K", marker="*")
plt.xlabel("Movie List")
plt.ylabel("Box-Office Collection in USD")
plt.title("Christopher Nolan Movie Earning")
plt.show()
```

```python
plt.scatter(x,y, color="K", marker="s")
plt.xlabel("Movie List")
plt.ylabel("Box-Office Collection in USD")
plt.title("Christopher Nolan Movie Earning")
plt.show()
```
